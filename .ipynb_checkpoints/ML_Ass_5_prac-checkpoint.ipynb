```python
import pandas as pd
```

```python
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('sales_data_sample.csv')
```

```python
df
```

```python
X = df[['QUANTITYORDERED','PRICEEACH']]
```

## Elbow plot

```python
no = []
inertia = []
for i in range(1,11):
    clust = KMeans(n_clusters=i).fit(X)
    no.append(i)
    inertia.append(clust.inertia_)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.plot(no,inertia)
plt.show()
```

### Elbow point is 4

## Clustering

```python
cluster = KMeans(n_clusters=4)
y_clusters = cluster.fit_predict(X)
```

## Plotting clusters

```python
cluster1 = [[],[]]
cluster2 = [[],[]]
cluster3 = [[],[]]
cluster4 = [[],[]]
def generate_values(X,y,n):
    for i in range(len(X)):
        if y[i]==0:
            cluster1[0].append(X['QUANTITYORDERED'].iloc[i])
            cluster1[1].append(X['PRICEEACH'].iloc[i])
        elif y[i]==1:
            cluster2[0].append(X['QUANTITYORDERED'].iloc[i])
            cluster2[1].append(X['PRICEEACH'].iloc[i])
        elif y[i]==2:
            cluster3[0].append(X['QUANTITYORDERED'].iloc[i])
            cluster3[1].append(X['PRICEEACH'].iloc[i])
        elif y[i]==3:
            cluster4[0].append(X['QUANTITYORDERED'].iloc[i])
            cluster4[1].append(X['PRICEEACH'].iloc[i])
```

```python
generate_values(X,y_clusters,4)
```

```python
plt.figure()
plt.scatter(cluster1[0],cluster1[1],c='blue',label='Cluster1')
plt.scatter(cluster2[0],cluster2[1],c='red',label='Cluster2')
plt.scatter(cluster3[0],cluster3[1],c='green',label='Cluster3')
plt.scatter(cluster4[0],cluster4[1],c='orange',label='Cluster4')
plt.scatter(cluster.cluster_centers_[:,0],cluster.cluster_centers_[:,1],c='black',label='Centroids')
plt.legend()
plt.xlabel('QUANTITYORDERED')
plt.ylabel('PRICEEACH')
plt.show()
```
