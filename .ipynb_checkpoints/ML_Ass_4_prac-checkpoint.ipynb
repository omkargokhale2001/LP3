```python
max_iter = 1000
curr_x = 2
precision = 0.0000001
delta_x = 999
no_of_iters = 0
lr = 0.01
x_values = []
y_values = []
```

```python
def get_slope(x):
    return 2*(x+3)
```

```python
def get_value(x):
    return (x+3)**2
```

```python
x_values.append(curr_x)
y_values.append(get_value(curr_x))
```

```python
def gradient_descent(no_of_iters,curr_x,delta_x):
    while delta_x>precision and no_of_iters<max_iter:
        prev_x = curr_x
        curr_x = curr_x - (lr*get_slope(prev_x))
        delta_x = abs(curr_x - prev_x)
        no_of_iters += 1
        x_values.append(curr_x)
        y_values.append(get_value(curr_x))
        print("## Iteration number " + str(no_of_iters))
        print("Current value of x is {}".format(curr_x))
        print("Current value of y is {}".format(get_value(curr_x)))
```

```python
gradient_descent(no_of_iters,curr_x,delta_x)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.plot(x_values,y_values)
plt.show()
```
